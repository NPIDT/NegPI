```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
"""These values were recuperated on the windows/features notebooks"""
precisions = pd.DataFrame([[83.09,80.43,85.25],
                           [84.63,82.64,83.08],
                           [83.26,83.14,83.14],
                           [84.76,83.14,83.81]],
                          index=['5','10','20','30'],
                          columns=['5','7','9'])

recalls = pd.DataFrame([[71.63,75.72,66.70],
                        [74.32,75.88,71.55],
                        [73.76,75.23,72.65],
                        [74.77,75.32,74.12]],
                        index=['5','10','20','30'],
                        columns=['5','7','9'])

fscores = pd.DataFrame([[76.85,77.95,74.77],
                        [79.04,79.03,76.72],
                        [78.17,78.96,77.43],
                        [79.42,79.42,78.61]],
                        index=['5','10','20','30'],
                        columns=['5','7','9'])
```

```python
fscore_labels = np.array([[str(x) for x in fscores.iloc[k]] for k in range(fscores.shape[0]-1,-1,-1)])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(fscores.iloc[::-1], annot=True, fmt='.2f', annot_kws={'size':'15'}, cmap='jet')
plt.xlabel('Window size', fontsize=18)
plt.ylabel('Top and bottom p% features used', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title('F1 scores heatmap', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(recalls.iloc[::-1], annot=True, fmt='.2f', annot_kws={'size':'15'}, cmap='jet')
plt.xlabel('Window size', fontsize=18)
plt.ylabel('Top and bottom p% features used', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title('Recalls heatmap', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(precisions.iloc[::-1], annot=True, fmt='.2f', annot_kws={'size':'15'}, cmap='jet')
plt.xlabel('Window size', fontsize=18)
plt.ylabel('Top and bottom p% features used', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title('Precisions heatmap', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(list(fscores.loc['5']), 'x-', label='F1-score for 5% features')
plt.plot(list(fscores.loc['10']), 'x-', label='F1-score for 10% features')
plt.plot(list(fscores.loc['20']), 'x-', label='F1-score for 20% features')
plt.plot(list(fscores.loc['30']), 'x-', label='F1-score for 30% features')
plt.xticks(range(3), ['5','7','9'], fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.xlabel('Window size', fontsize=16)
plt.ylabel('F1-score', fontsize=16)
plt.title('F1-score as a function of the window size', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(list(fscores['5']), 'x-', label='F1-score for 5-word windows')
plt.plot(list(fscores['7']), 'x-', label='F1-score for 7-word windows')
plt.plot(list(fscores['9']), 'x-', label='F1-score for 9-word windows')
plt.xticks(range(4), ['5','10','20','30'], fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.xlabel('Top and bottom % features words', fontsize=16)
plt.ylabel('F1-score', fontsize=16)
plt.title('F1-score as a function of the percentage of features words', fontsize=20)
plt.show()
```
